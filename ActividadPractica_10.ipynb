# **Guía del Laboratorio: Clasificación con el Dataset Iris**

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
````

````python
iris = load_iris()
X = iris.data[:, :2] # Extrae las dos primeras características: longitud y ancho del sépalo
y = iris.target     # Extrae las etiquetas de las especies

print("Shape de las características (X):", X.shape)
print("Shape de las etiquetas (y):", y.shape)
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Shape de X_train:", X_train.shape)
print("Shape de X_test:", X_test.shape)
print("Shape de y_train:", y_train.shape)
print("Shape de y_test:", y_test.shape)
````

Ahora, te explico la línea de código que acabamos de generar:

```python
model = LogisticRegression()
```

1.  **`model = LogisticRegression()`**
    *   **Clase:** `LogisticRegression`
        *   **Explicación:** Esta es una clase de la librería `sklearn.linear_model` que implementa el algoritmo de Regresión Logística. A pesar de su nombre, la Regresión Logística es un algoritmo de *clasificación*, no de regresión. Se utiliza para modelar la probabilidad de que una instancia de entrada pertenezca a una clase determinada.
        *   Cuando creamos `model = LogisticRegression()`, estamos inicializando una nueva instancia de este clasificador. En este punto, el modelo está "vacío" o "sin entrenar"; aún no ha aprendido nada de nuestros datos. Lo estamos preparando para que esté listo para el paso de entrenamiento.
    *   **Resultado:** La variable `model` ahora contiene un objeto `LogisticRegression`, listo para ser entrenado con nuestros datos de entrenamiento (`X_train`, `y_train`).

En esencia, esta línea de código le dice a Python: "Voy a usar el algoritmo de Regresión Logística para mi tarea de clasificación. ¡Prepárame una versión de ese algoritmo para que pueda empezar a trabajar con él!". Después de esto, el siguiente paso lógico es entrenar el modelo con los datos.

````python
model = LogisticRegression()
print("Modelo de Regresión Logística creado:", model)
````

````python
model.fit(X_train, y_train)
````

````python
y_pred = model.predict(X_test)
````

````python
accuracy = accuracy_score(y_test, y_pred)
print(f"Precisión del modelo: {accuracy:.2f}")
````

````python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(x=X_test[:, 0], y=X_test[:, 1], hue=y_test, palette='viridis', legend='full')
plt.title('Especies Reales (Conjunto de Prueba)')
plt.xlabel('Longitud del Sépalo')
plt.ylabel('Ancho del Sépalo')

plt.subplot(1, 2, 2)
sns.scatterplot(x=X_test[:, 0], y=X_test[:, 1], hue=y_pred, palette='viridis', legend='full')
plt.title('Predicciones del Modelo (Conjunto de Prueba)')
plt.xlabel('Longitud del Sépalo')
plt.ylabel('Ancho del Sépalo')

plt.tight_layout()
plt.show()
````
