```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {'Mes': ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"], "Ventas": [25, 30, 28, 35, 40, 50, 55, 60, 70, 80, 75, 90],}
data
```

```python
data2 = {"Estudiante": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah", "Ian", "Jack"], "Matemáticas": [85, 92, 78, 90, 88, 76, 95, 89, 82, 94], "Ciencias": [90, 85, 80, 88, 92, 75, 95, 87, 81, 93], "Historia": [78, 82, 88, 90, 85, 80, 92, 84, 79, 91]}
df_calificaciones = pd.DataFrame(data2)
display(df_calificaciones.head())
```

Now that we have the student grades in a DataFrame, let's perform some statistical analysis using NumPy to understand the distribution of grades in each subject.

```python
print("Análisis Estadístico de Calificaciones por Asignatura:")
for subject in ['Matemáticas', 'Ciencias', 'Historia']:
    grades = df_calificaciones[subject].to_numpy()
    mean_grade = np.mean(grades)
    std_dev_grade = np.std(grades)
    min_grade = np.min(grades)
    max_grade = np.max(grades)
    print(f"\n{subject}:")
    print(f"  Promedio: {mean_grade:.2f}")
    print(f"  Desviación Estándar: {std_dev_grade:.2f}")
    print(f"  Calificación Mínima: {min_grade}")
    print(f"  Calificación Máxima: {max_grade}")
```

Finally, let's visualize the average grades for each subject to get a quick overview.

```python
avg_grades_per_subject = df_calificaciones[['Matemáticas', 'Ciencias', 'Historia']].mean()

plt.figure(figsize=(8, 5))
plt.bar(avg_grades_per_subject.index, avg_grades_per_subject.values, color=['lightcoral', 'lightgreen', 'lightblue'])
plt.xlabel('Asignatura')
plt.ylabel('Calificación Promedio')
plt.title('Calificación Promedio por Asignatura')
plt.ylim(0, 100) # Assuming grades are out of 100
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
